```python
'''
5.开始训练模型
SVM分类器
'''
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import time
import numpy as np
import matplotlib.pyplot as plt
C_range = np.linspace(0.01, 30, 50)
test1 = []
test2 = []
for i in  C_range:
    start_time = time.time()
    #创建模型
    SVM = SVC(C=i,kernel='rbf',gamma='auto')
    #拟合数据
    SVM.fit(train_x_tfvec,train_y)
    #训练集
    pre_train_y = SVM.predict(train_x_tfvec)
    train_accuracy = accuracy_score(pre_train_y,train_y)
    test1.append(train_accuracy)
    test2.append(test_accuracy) 
plt.plot(C_range, test1, color = 'red', label = " max_depth_train")
plt.plot(C_range, test2, color = 'blue', label = " max_depth_test")
plt.legend()
plt.show()
```
